```python
import re
import pandas as pd
from bibtexparser.bwriter import BibTexWriter
from bibtexparser.bibdatabase import BibDatabase
```

```python
worddoc = "E:/UBOPAll.txt"
indexnum, authorlast, authorfirsti, yeardate, authlist = [],[],[],[],[]
title, journalbits, annotation, pubtype = [],[],[],[]
fmdesc ={}
dfdict={}
with open(worddoc, 'r') as file:
    text = file.read()
    if "\#endcard\n\#card":

        textlist = text.split("#endcard")
        for i in range(len(textlist)):
            #indexnum.append(textlist[i][0:textlist[i].find(' ')].strip('\n').strip('"').strip("#card"))
            indstring = textlist[i][0:textlist[i].find(' ',re.search("\d",textlist[i]).start())]
            indstring = indstring.replace('\n','').replace('"','').replace('#card','').strip(" ")
            indexnum.append(indstring)
            textsub = textlist[i][:textlist[i].find('\n',5)]
            firstbreak = textlist[i].find('\n',6)
            secondbreak = textlist[i].find('\n',firstbreak+1)
            thirdbreak = textlist[i].find('\n',secondbreak+2)
            title.append(textlist[i][firstbreak+1:secondbreak].strip("."))                     
            journalbits.append(textlist[i][secondbreak+1:thirdbreak].split(","))
            annotation.append(textlist[i][textlist[i].find("(Annotation)")+len("(Annotation)"):textlist[i].find("FORMATION")].strip("\n"))
            typebits = journalbits[i][0]
            if len(journalbits[i]) > 1:
                volume = journalbits[i][1].strip('v.')
            else:
                volume = ''
            if "Ph" or "hesis " in typebits:
                pubtype.append('Thesis')
            elif "United States Geological Survey" or "Utah Geological" in typebits:
                pubtype.append('Report')
            else:
                pubtype.append('Journal Article')
            g = [m.start() for m in re.finditer(',',textsub)]
            try:
                yrdtind = re.search('[1][8,9][0-9]{2}',textsub)
                authlist.append(textlist[i][textlist[i].find(' '):re.search(', [1][8,9][0-9]{2}',textlist[i]).start()].strip())
                #yeardate.append(textsub[textsub.find(', ',g[-1])+2:textsub.find(', ',g[-1])+6])
                yeardate.append(textsub[yrdtind.start():yrdtind.end()])
            except:
                yeardate.append(0)
                authlist.append(' ')
            authorlast.append(textlist[i][textlist[i].find(' '):textlist[i].find(',')].strip())
            authorfirsti.append(textlist[i][textlist[i].find(','):textlist[i].find(',',1)].strip())    
            flfindex = textlist[i].find('FORMATION          LOCATION                      FOSSILS\n')
            fmdesc[indexnum[i]] = textlist[i][flfindex:]
            print(indexnum[i])
            dfdict[indexnum[i]] = pd.DataFrame({'Index':int(indexnum[i]),'Authorlist':authlist[i],'Year':yeardate[i],
                                                "PubType":pubtype[i],"Vol":volume, "Journal":typebits },index=['Index'])
            
            
            
            
pubtype[120]
```

```python
            db = BibDatabase()
            db.entries = [
                {'journal': typebits,
                 'comments': annotation[i],
                 'pages': journalbits[i][-1].strip('p.').strip('pp.'),
                 'title': title[i],
                 'year': yeardate[i],
                 'volume': journalbits[i][1].strip('v.'),
                 'id': 'Paleo'+indexnum[i].zfill(4),
                 'author': authlist[i],
                 'type': pubtype[i]}]

            writer = BibTexWriter()
            with open('E:/bibtex.bib', 'a') as bibfile:
                bibfile.write(writer.write(db))
```
